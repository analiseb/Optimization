### 1 One-dimensional Case

$$ f(x)= x^{3}-2x+2 $$

#### 1.1. Plot this function

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2,2,100)
y = x**3 - 2*x + 2

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,y)
plt.show()
```

#### 1.2 Analytical computation of f'(x)=0

$$ f'(x) = 3x^{2} - 2 $$

$$ 0 = 3x^{2} - 2 $$

$$ 2 = 3x^{2} $$

$$ 2/3 = x^{2} $$

$$ \pm sqrt(2/3) = x $$

#### 1.4 Plot the second derivative


```python
# plot first and second derivative
y1 = 3*x**2-2
y2 = 6*x

fig2 = plt.figure()
ax = fig2.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(['first derivative', 'second derivative'])
```

>The second derivative at (-sqrt(2/3))<0 and is a maximum, conversely at (sqrt(2/3))>0 and is a minimum.

#### 2.2.3 Numerically compute an approximation of the points x* at which f(x*)=0

In order to find the local extrema of the two-dimensional function, we follow the methods as proposed in the document.

First, we calculate the magnitude of the gradient at each point.

```python
# 2.2.3 Numerically compute an approximation of the points x* at which f(x*)=0

# (a) evaluate ||f'(x)||^2 at the previuos range using a step of eg 0.005 or smaller
# create a matrix that stores all the latter values to analyze in next steps

# We have used matrixcalculus.org to analytically compute the gradient.
def z5_gradient(x,y):
    dx = y + 2*x*(4-2.1*x**2 + (1/3)*x**4) - 4.2*x**3 + (4*x**5)/3
    dy = x + 2*y*(4*y**2 - 4) + 8*y**3
    return dx**2 + dy**2

# Use step 0.005, so in every segment of size 1 we take 200 points.
# We can only use non-negative integers as matrix indexes, therefore:
#     - The interval [-2,2] becomes [0,799] 
#     - The interval [-1,1] becomes [0,399]
mat = np.zeros(shape=(800,400))
for i in range(0,800):
    for j in range (0, 400):
        result = z5_gradient((i-400)/200,(j-200)/200)
        mat[i,j] = result
```

```python
mat.shape
```

Then we find the values in which the magnitude of the gradient is zero.

```python
# (b) brute force minsearch within 8 neighboring points to find candidates

import pandas as pd
# Data frame to store resulting points
df_results = pd.DataFrame(columns = ['x', 'y', 'gradient_size'])

for i in range(1,799):
    for j in range(1,399):
        m0 = mat[i,j]
        m1 = mat[i-1,j-1]
        m2 = mat[i,j-1]
        m3 = mat[i+1,j-1]
        m4 = mat[i+1,j]
        m5 = mat[i+1,j+1]
        m6 = mat[i,j+1]
        m7 = mat[i-1,j+1]
        m8 = mat[i-1,j]
        if m0<m1 and m0<m2 and m0<m3 and m0<m4 and m0<m5 and m0<m6 and m0<m7 and m0<m8:
            # print('(i-400)/200='+str((i-400)/200)+ '    (j-200)/200=' + str((j-200)/200) + '    value=' + str(round(m0,4)))
            df_results = df_results.append({'x' : (i-400)/200, 'y' : (j-200)/200, 'gradient_size' : m0},
                                           ignore_index=True) 
```

#### (4) Compute Hessian of f(x1,x2)

#### What kind of information is the Hessian giving you?

>For the points where the gradient is zero, the eigenvalues of the hessian matrix indicate if a given point is a maximum, minimum, or saddle-point.

```python
# Hessian calculated analytically
def hessian_eigs(x,y):
    d2x = 2*(4 - 2.1*x**2 + (x**4)/3) - 8.4*x**2 + 8/3*x**4 - 3*4.2*x**2 + 20/3*x**4
    d2y = 2*(4*y**2 - 4) + 16*y**2 + 24*y**2
    H = [[d2x, 1], [1, d2y]]
    return H

# Calculate the hessian eigenvalues for each of the candidate points.
df_results['hessian_eigs'] = pd.Series(
    LA.eigvals(hessian_eigs(row.x, row.y))
    for index, row in df_results.iterrows()
)

# Label each point as max, min or saddle, depending on the sign of the eigenvalues.
def max_min_saddle(eigen1, eigen2):
    if eigen1>0 and eigen2>0:
        result = 'min'
    elif eigen1<0 and eigen2<0:
        result = 'max'
    else:
        result = 'saddle'
    return result

# Create max_min_saddle column in results table
df_results['max_min_saddle'] = pd.Series(
    max_min_saddle(row.hessian_eigs[0], row.hessian_eigs[1])
    for index, row in df_results.iterrows()
)

df_results
```
